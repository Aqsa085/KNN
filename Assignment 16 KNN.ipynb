##  Handle Outliers

```python
# Example: Cap outliers based on the IQR method
Q1 = df['hair'].quantile(0.25)
Q3 = df['hair'].quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
# Limit outliers
df = df[(df['hair'] >= (Q1 - 1.5 * IQR)) & (df['hair'] <= (Q3 + 1.5 * IQR))]
```

##  Choose an Appropriate Distance Metric and Value for K
##  Experiment with Different K Values

```python
from sklearn.model_selection import cross_val_score

# Check accuracy for different K values
k_values = range(1, 21)
accuracies = []
```

```python
k_values
```

```python

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    accuracies.append(scores.mean())
```

```python
scores
```

```python
# Plot the accuracies for different K values
plt.plot(k_values, accuracies)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Cross-Validated Accuracy')
plt.title('K vs Accuracy')
plt.show()
```

##  Evaluate the Classifier's Performance
##  Make Predictions and Evaluate

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
# Display the metrics
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

##  Visualize the Decision Boundaries
##  Visualize the Decision Boundary

```python
# Reduce to two dimensions for visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

```python
X_reduced
```

```python
# Train KNN on the reduced dataset
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_reduced, y)
```

```python
# Create mesh for plotting
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))
```

```python
xx, yy
```

```python
# Predict on the mesh
# Predict on the mesh using the PCA-transformed meshgrid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

# Reshape the result back to the meshgrid shape
Z = Z.reshape(xx.shape)
```

```python
Z
```

```python
# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, edgecolors='k', marker='o')
plt.title('KNN Decision Boundary')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

## Interview Questions:

## 1. What are the key hyperparameters in KNN?

1. K (Number of Neighbors): Number of neighbors considered for prediction.
    
2. Distance Metric: How distance is calculated (e.g., Euclidean, Manhattan).
    
3. Weights: Whether neighbors contribute equally or based on distance.
    
4. Algorithm: Method for finding neighbors (e.g., brute-force, KD-Tree).
    
5. p (Minkowski Distance): Power parameter; controls whether distance is Euclidean (p=2) or Manhattan (p=1).

## 2. What distance metrics can be used in KNN?

Here are the distance metrics used in KNN:

1. Euclidean Distance: Measures the straight-line distance between two points.
    
2. Manhattan Distance: Measures the distance by summing the absolute differences along each axis.
    
3. Minkowski Distance: A generalized distance metric; it includes both Euclidean and Manhattan as special cases.
    
4. Cosine Similarity: Measures the angle between two vectors, often used in text or high-dimensional data.
    
5. Hamming Distance: Used for categorical data, counts the number of different positions between two strings or binary vectors.
    
These metrics help determine how "close" or "far" points are from each other.
